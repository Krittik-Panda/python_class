### 1)

```python
s = input("Enter string: ")
vowels = "aeiouAEIOU"
count = 0
for ch in s:
    if ch in vowels:
        count += 1
print("Vowel count:", count)
```

### 2)

```python
str = input("enter the string :")
print(str)
reverse = ""
for ch in str:
    reverse = ch + reverse
print(f"reversed string : {reverse}")    
```

### 4)

```python
text = input("Enter a string: ")

freq = {}   

for ch in text:
    if ch in freq:
        freq[ch] += 1   
    else:
        freq[ch] = 1   

print(freq)
```
